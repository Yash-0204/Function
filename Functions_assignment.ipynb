Q.  Explain the concept of generators in Python and how they are defined?

    A generator is a special type of iterator in Python that allows you to iterate over a potentially large sequence of data, one item at a time, without storing the entire sequence in memory. This makes generators highly memory-efficient, especially for handling large datasets or when you need to generate values lazily.

    It is defined by

    Using a Generator Function (with the yield keyword)
    Using Generator Expressions (similar to list comprehensions but with parentheses).

Q. Explain the purpose and usage of the `map()` function in Python?

    The map() function in Python is a higher-order function that allows you to apply a given function to every item in an iterable (such as a list, tuple, etc.), returning an iterator that produces the results. It is a powerful tool for transforming or processing collections of data in a concise and functional way.

    Example: def square(x):
             return x ** 2
             numbers = [1, 2, 3, 4, 5]
             squared_numbers = map(square, numbers)   
             print(list(squared_numbers))  
             Output: [1, 4, 9, 16, 25]
